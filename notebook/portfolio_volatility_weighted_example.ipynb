```python
def get_symbol_from_scrapy_giant(stockid, opt, start=datetime.utcnow() - timedelta(days=90), end=datetime.utcnow()):
    # get hisstock default raw data 
    kwargs = {
        'opt': opt,
        'targets': ['stock'],
        'starttime': starttime,
        'endtime': endtime,
        'stockids': [stockid],
        'traderids': [],
        'base': 'stock',
        'callback': None,
        'limit': 1,
        'debug': True #only use for dbg db
    }
    st_panel, dbhandler = collect_hisframe(**kwargs)
    if st_panel.empty:
        return
    px = st_panel[stockid]
    rets = px[['close']].pct_change().dropna()
    #rets.index = rets.index.tz_localize("UTC")
    rets.columns = [stockid]
    return rets[stockid]
```

```python
# ret = 2317 , benchmark = 2330
starttime = datetime.utcnow() - timedelta(days=90)
endtime = datetime.utcnow()

DF_2317 = get_symbol_from_scrapy_giant('2317', 'twse', starttime, endtime)
DF_2330 = get_symbol_from_scrapy_giant('2330', 'twse', starttime, endtime)
DF_2317.name = '2317'
DF_2330.name = '2330'
stocks_str = ['2317', '2330']
stocks = [DF_2317, DF_2330]
stocks_df_na = pd.DataFrame(stocks).T
stocks_df_na.columns = stocks_str
stocks_df = stocks_df_na.dropna()
```
